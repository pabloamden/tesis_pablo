## Script para corregir un offset en el tiempo de inicio del archivo rttm derivado del .eaf


```python
import pandas as pd

# Load the RTTM file into a DataFrame
input_file = "casos/juliaa-a1-nsm/juliaa-a1-nsm.rttm"
output_file = "casos/juliaa-a1-nsm/juliaa-a1-nsm_modified.rttm"

# Read the RTTM file into a DataFrame, assuming space-separated columns
df = pd.read_csv(input_file, sep="\s+", header=None, names=["Type", "File ID", "Channel", "Start Time", "Duration", "Orthography1", "Orthography2", "Speaker ID", "Turn Number"])

# Remove the first row
df = df.iloc[1:]

# Subtract 18000 from the "Start Time" column
df["Start Time"] = df["Start Time"] - 18000

# Save the modified DataFrame back to an RTTM file with identical formatting
with open(output_file, "w") as f:
    for _, row in df.iterrows():
        # Format each row to match the original RTTM structure
        line = f"{row['Type']}\t{row['File ID']}\t{int(row['Channel'])}\t{row['Start Time']:.2f}\t{row['Duration']:.2f}\t{row['Orthography1']}\t{row['Orthography2']}\t{row['Speaker ID']}\t{int(row['Turn Number'])}\n"
        f.write(line)

print(f"Modified RTTM file saved to {output_file}")

```
