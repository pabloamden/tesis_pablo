```python
from pydub import AudioSegment
from pydub.silence import detect_nonsilent

# Ruta al archivo de audio (puedes cambiarlo según la ubicación de tu archivo)
archivo_audio = "C:/Users/pablo/OneDrive/Escritorio/Audios de archivos llamativos/amparof-a2-nsm.wav"

def detectar_variaciones_audio(archivo_audio):
    # Cargar el archivo de audio
    audio = AudioSegment.from_file(archivo_audio)

    # Obtener los últimos 20 segundos del audio
    duracion_total = len(audio)
    duracion_ultimos_20s = 20 * 1000  # 20 segundos en milisegundos
    inicio_ultimos_20s = duracion_total - duracion_ultimos_20s
    audio_ultimos_20s = audio[inicio_ultimos_20s:]

    # Detectar variaciones en los niveles de señal o silencio digital
    segmentos_silencio = detect_nonsilent(audio_ultimos_20s, min_silence_len=1000, silence_thresh=-50)

    if len(segmentos_silencio) > 0:
        print("Se detectó silencio digital en los últimos 20 segundos de audio.")
    else:
        print("No se detectó silencio digital en los últimos 20 segundos de audio.")


detectar_variaciones_audio(archivo_audio)
```
