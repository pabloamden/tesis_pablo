```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import librosa
from tqdm import tqdm
```

```python
def detectar_amplitud_cero(archivos_audio, ventana_size=22050):
    paths_con_silencio = []
    paths_ok = []
    for archivo_audio in tqdm (archivos_audio):
        y, sr = librosa.load(archivo_audio)

        tiene_silencio = False
        for i in range(0, len(y) - ventana_size + 1, ventana_size):
            ventana = y[i:i+ventana_size]

            if np.all(ventana == 0):
                paths_con_silencio.append(archivo_audio)
                tiene_silencio = True
                break  # Salimos del bucle interno para el archivo actual
                
        
        if not tiene_silencio:
            paths_ok.append(archivo_audio)
              
                
    return paths_con_silencio, paths_ok
   

columna_audio = 'path_server'
df = pd.read_csv('test_errores.csv') 

archivos_audio = df[columna_audio].tolist()

resultado = detectar_amplitud_cero(archivos_audio)
 
if resultado:
    print("Se encontró una ventana de amplitud cero en los siguientes archivos:")
    for archivo in resultado:
        print(archivo)
else:
    print("No se encontró una ventana de amplitud cero en ningún archivo.")
```
