```python
import speach.elan
import pandas as pd
```

```python
from speach import elan
eaf = elan.read_eaf('alma-a1-nsb.eaf')
eaf
```

```python
for tier in eaf:
    print(f"{tier.ID} | Participante: {tier.participant} | Tipo: {tier.type_ref}")
    for ann in tier:
        print(f"{ann.ID.rjust(4, ' ')}. [{ann.from_ts.ts} -- {ann.to_ts.ts}] {ann.text}")
```

```python

# Crea una lista con los tiers
tiers_list = [tier.ID for tier in eaf]
tiers_list
```
